# 1. Series
1차원 데이터(정수, 실수, 문자열 등)

## Series 객체 생성
예) 1월부터 4월까지 평균 온도 데이터

```python
temp = pd.Series([-20,-10,10,20]) #list형태로 전달
print(temp)
```

#9. 데이터 수정

```python
 import pandas as pd
 data = { #dictionary type. key: string, value = list
    "이름" : ["석한별", "김도훈", "석별한", "별한이", "훈도김", "김또훈", "김둔호", "카악이"],
    "학점" : [4.5, 4.5, 4.3, 4.2, 3.6, 3.2, 2.7, 2.4],
    "주력언어" : ["R", "Java", "Python", "Python", "Javascript", "React", "Python", "Java"]
 }

df=pd.DataFrame(data)
df["결과"] = 'Fail'
df
```

```python
df.loc[df["학점"] > 3.8, '결과'] = "Pass"
df
```

```python
df.drop(index = df[(df["결과"] == "Pass")].index)
```

```python
def plusScore(value):
  return value + 0.5

df['학점'] = df['학점'].apply(plusScore)
df
```

#10. 그룹화

```python
 import pandas as pd
 data = { #dictionary type. key: string, value = list
    "이름" : ["석한별", "김도훈", "석별한", "별한이", "훈도김", "김또훈", "김둔호", "카악이"],
    "학점" : [4.5, 4.5, 4.3, 4.2, 3.6, 3.2, 2.7, 2.4],
    "주력언어" : ["R", "Java", "Python", "Python", "Javascript", "React", "Python", "Java"]
 }

df=pd.DataFrame(data)
df["결과"] = 'Fail'
df.loc[df["학점"] > 3.8, '결과'] = "Pass"
df
```

```python
df.groupby("결과")
```

```python
df.groupby("결과").get_group("Pass")
```

```python
df.groupby("결과").get_group("Fail").mean()
```

```python
df.groupby(["주력언어", "결과"]).mean()
```
